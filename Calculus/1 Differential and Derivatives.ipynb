Suppose that $f(x)$ is continuous on $[a, b]$ and let $P = (x_0, f (x_0))$
and $Q = (x_0 + h, f (x_0 + h))$ on the graph of $f (x)$:



Then the slope of $\overset{\longleftrightarrow}{P Q}$, secant line passing
through $P$ and $Q$, is

$$
\large{\mathbf{m_{\overset{\leftrightarrow}{PQ}} =  \frac{f (x_0 + h) - f (x_0)}{h} }}
$$

called *Newtonian quotient* of $f (x)$ at $x = x_0$. If the limit of $m_{\overset{\leftrightarrow}{PQ}}$, by moving
$Q$ towards $P$ along the graph of $f(x)$, (i.e. $h$
approaches $0$), exists, 

$$
\large{\mathbf{m = \lim_{Q\to P}m_{\overset{\leftrightarrow}{PQ}} =\lim\limits_{h \rightarrow 0} \frac{f (x_0 + h) - f (x_0)}{h} }}
$$
it is called the derivative of $f (x)$ at $x = x_0$,
i.e. it is the slope of tangent line of $f (x)$ passing through $(x_0, f
(x_0))$, and denoted as $f' (x_0)$. The process for finding derivative is
called differentiation.

Derivative
---

Suppose that $f(x)$ iss defined on $D\subset \mathbb{R}$. $f(x)$ is called diffwerentiable if $f'(x)$ exists for any $x_0$ in $D$.

The derivative of $f(x)$ is defined as follows:
$$f'(x)=\lim_{x\to a}\frac{f(x)-f(a)}{x-a}=\lim_{h\to0}\frac{f(a+h)-f(a)}{h}$$
where the limit exists for $a\in D$.

```python
# differtiation_d
def differentiation_d(f_,x_,val=""):
    fu=str(simplify(diff(f_,x_)))
    if val=="":
       text="\\begin{eqnarray}"      
       text=text+"\large{f'(%s)}&=&\large{\left(%s\\right)'}\cr" %(x_,latex((f_)))
       text=text+"&=&\large{\lim_{h\\to0}\\frac{f(%s+h)-f(%s)}{h}}\cr" %(x_,x_)
       fh=f_.subs({x_:x_+h})
       fhh=simplify(fh-f_)
       text=text+"&=&\large{\lim_{h\\to0}\\frac{%s-%s}{h}}\cr" %(latex((fh)),latex(f_)) 
       text=text+"&=&\large{\lim_{h\\to0}\\frac{%s}{h}}\cr" %(latex((fhh))) 
       #
       text=text+"&=&\large{%s}\end{eqnarray}" %(latex(eval(fu)))
    else:
       #c_=latex(val)
       f_val=f_.diff(x_).subs({x:val})
       #text="$$\large{f'\left(%s\\right)=\left.\left(%s\\right)'\\right|_{x=%s}=\left.%s\\right|_{x=%s}}=%s$$" \
       #      %(latex(val),latex((f_)),latex(val),latex(eval(fu)),latex(val),latex(f_val))
    
       text="\\begin{eqnarray}"      
       text=text+"\large{f'(%s)}&=&\large{\left.\left(%s\\right)'\\right|_{x=%s}}\cr" %(val,latex((f_)),latex(val))
       text=text+"&=&\large{\lim_{h\\to0}\\frac{f(%s+h)-f(%s)}{h}}\cr" %(val,val)
       fh=f_.subs({x_:val+h})
       f_0=f_.subs({x_:val})
       fhh=simplify(fh-f_0)
       text=text+"&=&\large{\lim_{h\\to0}\\frac{%s-%s}{h}}\cr" %(latex((fh)),latex(f_0)) 
       text=text+"&=&\large{\lim_{h\\to0}\\frac{%s}{h}}\cr" %(latex((fhh))) 
       #
       text=text+"&=&\large{%s}\end{eqnarray}" %(latex(f_val))
    return Latex(text)
```

Theorem
---
If $f(x)$ is differentiable at $x=a$, then $f(x)$ is continuous at $x=0$. But the converse migh be not held.

Proof
---
\begin{eqnarray}
  \lim\limits_{x\to a}|f(x)-f(a)|&=& \lim\limits_{x\to a}\left|\frac{f(x)-f(a)}{x-a}\right|\cdot |x-a|\\
  &=& |f'(a)|\cdot0\\
  &=& 0
\end{eqnarray}
This means $\lim\limits_{x\to a}f(x)=f(a)$, i.e. $f(x)$ is continuous at $x=a$.

**Counter-Example, ontinuous but not differentiable**

Consider $f(x)=|x|$, for $x\in\mathbb{R}$. $f(x)$ is continuous at $x=0$ but not diffenertiable:

1. $$\lim_{x\to0}|x|=0$$
2. $$\lim_{x\to0+}\frac{|x|-0}{x}=\lim_{x\to0+}\frac{x}{x}=1, \lim_{x\to0-}\frac{|x|-0}{x}=\lim_{x\to0-}\frac{-x}{x}=-1$$

Neither $x^{1/3}$ nor $x^{2/3}$ are not differentiable at $x=0$; the derivative of  the former at $x=0$ is infinite (i.e. non-exist), and the latter has a edge on its graph at the origin.

Suppose that
$$ f (x) = \left\{ \begin{array}{l}
       x \sin (1 / x) \text{, if } x \neq 0,\\
       0, \text{, if } x = 0
     \end{array} \right. $$
  Then $f (x)$ is continuous for any real number by the squeeze value theorem.
  And the the derivative at $x = 0$, is as follows:
  \begin{eqnarray*}
    f' (0) & = & \lim_{h \rightarrow 0} \frac{f (h) - f (0)}{h}\\
    & = & \lim_{h \rightarrow 0} \frac{h \sin (1 / h)}{h}\\
    & = & \lim_{h \rightarrow 0} \sin (1 / h)
  \end{eqnarray*}
  This concludes that the limit fails to exist.


Example
---
The derivative of sine function is derived as follows:
\begin{eqnarray*}
 (\sin x)' & = & \lim_{h \rightarrow 0} \frac{\sin (x + h) - \sin x}{h}\\
 & = & \lim_{h \rightarrow 0} \frac{\sin x \cos h + \cos x \sin h - \sin x}{h}\\
 & = & \lim_{h \rightarrow 0} \left( \sin x \frac{\cos h - 1}{h} + \cos x \frac{\sin h}{h} \right)\\
 & = & \cos x
\end{eqnarray*}

since 
$$\lim\limits_{h \rightarrow 0} \frac{\sin h}{h} = 1,
  \lim\limits_{h \rightarrow 0} \frac{\cos h - 1}{h}=0.$$

2.2 Rules of Differentiation, 2.3., (Product and Quotient Rules)
---


1. $(c)' = 0$ for any constant $c$.
2. $(x^n)' = n x^{n - 1}$ for $x > 0$ and $n \in \mathbb{R}$. Here, a
    example for $n = - 1$:

\begin{eqnarray*}
 \left(\frac{1}{x}\right)' & = & \lim_{h \rightarrow 0} \frac{\frac{1}{x + h} -
 \frac{1}{x}}{h}\\
 & = & \lim_{h \rightarrow 0} \frac{x - (x + h)}{x h (x + h)}\\
 & = & \frac{- 1}{x^2}
\end{eqnarray*}

3. $(f \pm g)' = f' \pm g'$ ,    
4.  $(c \cdot f)' = c\cdot f'$ where $c\in\mathbb{R}$; $(f \cdot g)' = f' g + g' f$,
5. $(f / g)' = (f' g - g' f) / g^2$ suppose that $g (x) \neq 0$ for any
    $x$,
6. (Power Rule) $(x^r)' = r x^{r - 1}$ for $r > 0$,
7. (Chain's Rule) $(f \circ g)' = (f (g (x)))' = f' (g (x)) g' (x)$  
$$
 (f (g (x)))' = \frac{d f (g (x))}{{d\not{g (x)}}} \cdot
     \frac{{{d\not{g (x)}}}}{d x}
$$     
 The following diagram says everything about chain's rule:

\begin{eqnarray*}
      \begin{array}{ccccc}
        &  &  &  & \\
        & g(x) &  & f(u) & \\
        x & \longrightarrow & u=g(x) & \longrightarrow & f(g(x))\\
        & \downarrow &  & \downarrow & \\
        & \color{brown}{g'(x)} & \times & \color{brown}{f'(u)} & \\
        &  & \downarrow &  & \\
        &  & \color{brown}{f'(g(x))g'(x) }&  & 
      \end{array} &  & 
\end{eqnarray*}

8. $(\sin x)' = \cos x$, $(\cos x)' = - \sin x$, $(\tan x)' = \sec^2 x$, $(\cot x)' = - \csc^2 x$, $(\sec x)' = \sec x \tan x$, $(\csc x)' = - \csc x \cot x$;
9. (Logarithmic Rule):
   - $(\ln x)' = (\log_e x)' = 1 / x$, where $\ln (\bullet)$ means
      logarithm with base $e \cong 2.71428 \cdots$ where $e$, Euler number, is an irrational
      number;
   - $(\log_a x)' = 1 / (x \ln a)$ where $a > 0$;
10. (Exponential Rule):
    - $e$, the Euler number:
      $$e=\lim_{n\to\infty}\left(1+\frac{1}{n}\right)^n$$
      
    - $(e^x)' = e^x$;
    - $(a^x)' = a^x \ln a$, where $a > 0$ and $a \neq 1$.

```python
# differtiation
def differentiation(f_,x_,val=""):
    fu=str(simplify(diff(f_,x_)))
    if val=="":
       text="$$\large{f'(%s)=\left(%s\\right)'=%s}$$" %(x_,latex((f_)),latex(eval(fu)))
    else:
       c_=latex(val)
       f_val=f_.diff(x_).subs({x:val})
       text="$$\large{f'\left(%s\\right)=\left.\left(%s\\right)'\\right|_{x=%s}=\left.%s\\right|_{x=%s}}=%s$$" \
             %(latex(val),latex((f_)),latex(val),latex(eval(fu)),latex(val),latex(f_val))
    return Latex(text)
```

Higher-order Derivatives
---
Suppose that $f(x)$ is smooth function; the second order derivative is defined as follows:
$$ f'' (x) = \frac{d^2}{d x^2} f (x)  = D_x^2 f (x) =\frac{d}{d x} f'(x) $$
And the more higher order derivatives, ($n$-th)-order for instance, are defined by the recurive formula:
\begin{eqnarray}
 f^{(3)}(x)&=& \frac{d^3}{d x^3} f (x)  = D_x^3 f (x)=  \frac{d}{d x} f'' (x) \\
 &\vdots&\\
 f^{(n)} (x) &=& \frac{d^n}{d x^n} f (x)  = D_x^n f (x)=  \frac{d}{d x} f^{(n-1)} (x) 
\end{eqnarray} 


Example
---
Investigate the derivatives of $f (x) = x^2 \sin x$.
  
**Sol:**
  
 The first order of derivative is:
\begin{eqnarray*}
 f' (x) & = & (x^2 \sin x)'\\
 & = & (x^2)' \sin x + x^2 (\sin x)'\\
 & = & 2 x \sin x + x^2 \cos x
\end{eqnarray*}
And the second order derivative is:
\begin{eqnarray*}
  f'' (x) & = & (x^2 \sin x)'' \\
  & = & (2 x \sin x + x^2 \cos x)'\\
  & = & (2 x)' \sin x + 2 x (\sin x)' + (x^2)' \cos x + x^2 (\cos x)'\\
  & = & 2 \sin x + 2 x \cos x + 2 x \cos x - x^2 \sin x\\
  & = & 2 \sin x + 4 x \cos x - x^2 \sin x
\end{eqnarray*}
Exactly, the last result is equal to the following:
$$
C (2, 0) (x^2)'' \sin x + C (2, 1) (x^2)' (\sin x)' + C (2, 2) x^2 (\sin
     x)'' 
$$
where $C (n, k) = n! / k! (n - k)$! and $n! = 1 \cdot 2 \cdot 3
  \cdots n$, the factorial of $n$. 
  
This relation can be extended to the more higher differentiation:

\begin{eqnarray*}
  f^{(n)} & = & (x^2 \sin x)^{(n)}\\
  & = & C (n, 2) (x^2)'' (\sin x)^{(n - 2)} + C (n, 1) (x^2)' (\sin x)^{(n- 1)} + C (n, 0) x^2 (\sin x)^{(n)}\\
  & = & n (n - 1) (\sin x)^{(n - 2)} + 2 n x (\sin x)^{(n - 1)} + x^2 (\sin x)^{(n)}
\end{eqnarray*}

By the last result, the derivative of $10$-th order is
$$ f^{(10)} (x) = 90 \sin x + 20 x \cos x - x^2 \sin x $$

2.5 Implicit Differentiation
---

Definition
---
 $y=f(x)$ is called `function` or `explicit function` prcisely; $y$ is called implicit function of $x$ if $F(x,y)=0$.

Locally, implicit function is explicit function. In other words,
$$ F(x,y)=0\Rightarrow F(x,y(x))=0$$

Thus, we can get the $y'(x)$ by differentiating both sides of equation with respect to $x$ and solving by the equation:
$$ \frac{d}{dx}F(x,y(x))=0$$

Example
---
The circle, $x^2 + y^2 = r^2$, is a famous known equation. And how do we find
the tangent line at the point on the circle? Implicit differentiation says
that $y$ can be treated as function of $x$ locally, i.e. $y = y (x)$. Then

\begin{eqnarray*}
  & x^2 + (y (x))^2 = r^2 & \\
  \Longrightarrow & [x^2 + (y (x))^2]' = [r^2]' & \text{ by differentiating both sides}\\
  \Longrightarrow & 2 x + 2 y (x) y' (x) = 0 & \\
  \Longrightarrow & y' (x) = - \frac{x}{y (x)} & \text{ if }y (x) \neq 0
\end{eqnarray*}

This means that $y' (x)$ depends on both $x$ and $y$. For example,
- At $(x, y) = (1 / \sqrt{2}, 1 / \sqrt{2})$, 
  $y' (x) = - \sqrt{2} /  \sqrt{2} = -1$;
- at $(x, y) = (- 3 / 5, 4 / 5)$, $ y' (x) = - [(- 3 / 5) / (4
  / 5)] =  3 / 4$

Visualization Implicit Fuctions
---
To visualize the implict function, just plot every pair of $(x,y)$'s where $x\in D(f)$ and $y=y(x)$. Sympy avails the function, plot_implicit()$, to visualize graph of the implicit funtions.

```python
from sympy.plotting import plot_implicit
from sympy import  symbols, Eq, solve

# While ultize Implicit plot function, x,y have to be treated as independent variables; acclaim them again:
x, y = symbols('x y')

def ImplicitPlot(express):
    eq = Eq(express,0)
    plot_implicit(eq,(x, -6, 6), (y, -4, 4), adaptive=False)
```

2.6 Derivatives of Inverse  Functions
---

Suppose that $y=y(x)$ for $x\in D\subset\mathbb{R}$. $g$ is called inverse of $f(x)$ if 
$$g(f(x))=x\text{ and } f(g(x))=x$$
And $g(x)$ is denoted as $f^{-1}$.

Example
---
1. $g(x)=\sqrt x$ is the inverse of $f(x)=x^2$ for $x\ge0$, since:
$$g(f(x))=\sqrt{x^2}=x\text{ and }f(g(x))=\left(\sqrt x \right)^2=x$$
But $g(x)$ is not inverse $x^2$ for $x\in\mathbb{R}$.
2. $\ln x$  for $x>0$ is the inverse of $\exp(x)$.

THeorem
---
Inverse of $f(x)$ exists if $f(x)$ is one to one, i.e. for every $y$ in range of $f(x)$, there is one and only one of $x$ in domain of $f(x)$ and $y=f(x)$.

Note
---
The graph of inverse of $f(x)$ is symmetry to the the graph of $f(x$) with repect to $y=x$. 

Inverse Differentiation
---
Suppose that the inverse of $y=f(x)$ exists and is equal to $x=f^{-1}(y)$. Then
$$\frac{d x}{dy}=\frac{1}{dy/dx}$$
i.e.
$$\left(f^{-1}(y)\right)'=\frac{1}{f'(x)}$$

Example 
---
Suppose that $f(x)=\frac{x^3}{4}+2x-1$. Find the derivative of $f^{-1}$ for $(x,y)=(2,5)$.
1. $f(x)$ is 1-1 since for $x_1\ne x_2$:
\begin{eqnarray}
f(x_2)-f(x_1)&=&\frac{x_2^3}{4}+2x_2-1-\frac{x_1^3}{4}-2x_1+1\\
               &=& \frac{1}{4}(x_2^3-x_1^3)+2(x_2-x_1)\\
               &=&(x_2-x_1)\left(\frac{1}{4}(x_2^2+x_1x_2+x_1^2)+2\right)\ne0
\end{eqnarray} 
$f^{-1}$ exists.
2. $f(2)=5$ and $f^{-1}(5)=2$.
3. $$f'(2)=\left.\frac{3x^2}{4}+2\right|_{x=2}=5$$
3. $$\left(f^{-1}\right)(5)=\frac{1}{f'(2)}=\frac{1}{5}$$
It is not necessary to find out the exact form of inverse to evaluate the derivative of inverse.

2.7 Newton's Method
---

Newton's method is a root finding method that uses linear approximation. In particular, we guess a solution  $x=c$ of the equation $f(x)=0$, compute the linear approximation of  $f(x)$ at $x=c$ and then find the $X$-intercept of the linear approximation.

Suppose that $f(x)$ is differentiable around its zero root, $f(c)=0$. Newton methods avails the scheme to find out the zero root numerically.

1. Find the initial point around $c$, says $x_0$; then we can approximate  by the tangent line at $x_0$  and compute the $X$-intercept of the tangent line. The equation of the tangent line at $x_0$ is
$$y = f'(x_0)(x - x_0) + f(x_0)$$
2. The $X$-intercept is the solution  of the equation:
$$0 = f'(x_0)(x_1 - x_0) + f(x_0)$$
and we solve for $x_1$:
$$x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}$$

3. Proceed the recursive formula:
$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$
4.$\{x_n\}$ converges to a solution of the equation.

```python
def newton(f,Df,x0,epsilon,max_iter):
    '''Approximate solution of f(x)=0 by Newton's method.

    Parameters
    ----------
    f : function
        Function for which we are searching for a solution f(x)=0.
    Df : function
        Derivative of f(x).
    x0 : number
        Initial guess for a solution f(x)=0.
    epsilon : number
        Stopping criteria is abs(f(x)) < epsilon.
    max_iter : integer
        Maximum number of iterations of Newton's method.

    Returns
    -------
    xn : number
        Implement Newton's method: compute the linear approximation
        of f(x) at xn and find x intercept by the formula
            x = xn - f(xn)/Df(xn)
        Continue until abs(f(xn)) < epsilon and return xn.
        If Df(xn) == 0, return None. If the number of iterations
        exceeds max_iter, then return None.

    Examples
    --------
    >>> f = lambda x: x**2 - x - 1
    >>> Df = lambda x: 2*x - 1
    >>> newton(f,Df,1,1e-8,10)
    Found solution after 5 iterations.
    1.618033988749989
    '''
    xn = x0
    for n in range(0,max_iter):
        fxn = f(xn)
        if abs(fxn) < epsilon:
            print('Found solution after',n,'iterations.')
            return xn
        Dfxn = Df(xn)
        if Dfxn == 0:
            print('Zero derivative. No solution found.')
            return None
        xn = xn - fxn/Dfxn
        print(xn)            
    print('Exceeded maximum iterations. No solution found.')
    return None
```

```python
f = lambda x: x**3 - x**2 - 1
Df = lambda x: 3*x**2 - 2*x
approx = newton(f,Df,1,1e-10,10)
print(approx)
```
