Example
---
  Evaluate the integral $$\Large{\int^{\infty}_0 e^{- x^2} d x}$$
  
**Sol:** Let $I = \int^{\infty}_0 e^{- x^2} d x$. Then $I =
  \int^{\infty}_0 e^{- y^2} d y$ by changing the dummy variable $x$ into $y$.
  Consider the product:

\begin{eqnarray*}
    I^2 & = & I \cdot I\\
    & = & \int^{\infty}_0 e^{- x^2} d x \cdot \int^{\infty}_0 e^{- y^2} d y\\
    & = & \int^{\infty}_0 \int^{\infty}_0 e^{- x^2 - y^2} d x d y\\
    & = & \int^{\pi / 2}_0 \int^{\infty}_0 e^{- r^2} \cdot r d r d \theta\\
    & = & \int^{\pi / 2}_0 \frac{1}{2} d \theta = \frac{\pi}{4}
  \end{eqnarray*}

\begin{eqnarray*}
    D & = & \{ (x, y) | 0 \leqslant x, y < \infty \}\\
    & = & \{ (r, \theta) | 0 \leqslant r < \infty, 0 \leqslant \theta
    \leqslant \pi / 2 \}
  \end{eqnarray*}

i.e. the whole first quadrant. This implies $I = \Large{\frac{\sqrt{\pi}}{2}}$ .

**2.** To prove

$$ \Large{\int^{\infty}_{- \infty} \frac{1}{\sqrt{2 \pi} \sigma} e^{\frac{- (x -
     \mu)^2}{2 \sigma^2}} d x = 1} $$

change the variable by substitution of 
$$\Large {t = \frac{x - \mu}{\sqrt{2} \sigma}
  \text{ and }d t = \frac{d x}{\sqrt{2} \sigma}}$$. 
  
  Also

$$ \Large{x \left| \underset{- \infty}{\overset{\infty}{\uparrow}}\right. \Longrightarrow t =
     \frac{x - \mu}{\sqrt{2} \sigma} \left| \underset{-
     \infty}{\overset{\infty}{\uparrow}} \right.}$$

Then

\begin{eqnarray*}
   \Large{ \int^{\infty}_{- \infty} \frac{1}{\sqrt{2 \pi} \sigma} e^{\frac{- (x -
    \mu)^2}{2 \sigma^2}} d x }& = & \int^{\infty}_{- \infty}
    \frac{1}{\sqrt{\pi}} e^{- t^2} d t\\
    & = & \frac{1}{\sqrt{\pi}} \cdot \sqrt{\pi} = 1
  \end{eqnarray*}

**3.** Similarly, we can also calculate

  \infty} \frac{x}{\sqrt{2 \pi} \sigma} e^{\frac{- (x - \mu)^2}{2 \sigma^2}} d
  x = \mu}$$
  Using the substitution, $$t = \frac{x -

\begin{eqnarray*}

Conclusion
---
1. **Shift invariance**: for any $\mu\in\mathbb{R}$
$$\Large{\int^{\infty}_{- \infty} e^{-(x-\mu)^2}dx= \int^{\infty}_{- \infty} e^{-x^2}dx=\sqrt{\pi}}$$
2. **Scaling**: or any $a\in\mathbb{R}$
$$\Large{\int^{\infty}_{- \infty} e^{-a^2x^2}dx=\frac{\sqrt \pi}{|a|}}$$
