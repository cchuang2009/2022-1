<b>Victor Klee</b>, 1973, asked:

      If they must, together, view every wall and space in an art gallery, how many guards are necessary? (representing the gallery as a simple polygon with n sides)

And <b>Chvátal</b>, 1975, proved this theorem answering Klee’s Problem:

<b>Theorem.</b>  <Big><b>$\mathbf{\color{blue}{\lfloor \frac{n}{3}\rfloor}}$</b></Big> guards are sometimes necessary and always sufficient to guard an art gallery with `n` walls.

Original proof was lengthy but Steve Fisk, 1978, used the coloring of triangle vertices to prove this theorem vibrantly and simply!

<b>Fact</b>: the <i>key</i>, Triangle is convex, i.e. the connected line segment of any two points in a set is also included in the set:   

     For any triangulation, the vertices of P can be colored with red, blue, and green in such a way that each triangle has a vertex of each color.

     Since there are n vertices, there is one color that is used for at most ⌊(n/3)⌋ vertices.

<b>Triangulation</b>

     Colored with three colored sets of potential guards. Assign any vertex one color, and an adjacent vertex another color. All the rest of the colors are determined.

     Either the 3 Green guards or the 3 Red guards are sufficient.
     
  

Reference
---
S. Fisk, A Short Proof of Chvátal’s Watchman Theorem, J. Combin. Theory B 24 (1978), 374
