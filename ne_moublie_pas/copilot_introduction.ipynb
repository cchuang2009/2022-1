What Copilot is 
---

There are tons of free/non-free software hosted in github, and official github provides a plugin, based on these sources, which helps uses to write a `neat` code, including

1. auto-generate the code in your editor (VS-code);
2. You can choose more options by yourself if U are not satisfied. More details:
```
[Tab]: Accept inline code suggestion
[ESC]: Dismiss inline code suggestion 
Alt + ] or Option (⌥) + ]: Show next suggestion 
Alt + [ or Option (⌥) + [: Show previous suggestion
Alt + \ or Option (⌥) + \: Trigger suggestion 
Ctrl + Enter: Open ten suggestions in a separate pane 
```

3. Not only Python supported, most the computer languages are also supported. ( this is enough for you to become a senior software engineer in the future).
