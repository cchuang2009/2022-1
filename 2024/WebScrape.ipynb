```python
df.info()
```

```python
from pyecharts.charts import Pie
from pyecharts import options as opts
```

```python
countries=df['Country'].values
ExportValue=df['Top Export Value (USD Billions)'].values
dataset=[list(z) for z in zip(countries,ExportValue)]
```

```python
c = (
    Pie()
    .add("CO2 Emitting",dataset[:20],
         radius=["45%", "75%"],
         center=["45%", "50%"],
         rosetype="radius",
        )
    # set title and legend
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Exports in Asia",pos_left="30%",),
        legend_opts=opts.LegendOpts(type_="scroll", pos_left="85%", orient="vertical"),
    )
    # set ticks
    .set_series_opts(label_opts=opts.LabelOpts(formatter=" {b}:  {c} B$"))
    )

c.render()
```
