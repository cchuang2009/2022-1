Setup GitHub login Password (SSH), 設定 GitHub 登錄密碼 (ssh)
---
1. By [Groq](groq.com)

```
Step 1: Generate a new SSH key
---
  - Open the Terminal app on your Mac.
  - Run the command ssh-keygen -t ed25519 to generate a new SSH key pair.
  -  Follow the prompts to set a passphrase for your key (optional, but recommended for 
    added security), or by click.
  - Save the key pair to a file (e.g., ~/.ssh/id_ed25519 and ~/.ssh/id_ed25519.pub).

Step 2: Copy the public key
---
 - Open the id_ed25519.pub file in a text editor (e.g., TextEdit).
 - Copy the entire contents of the file.

Step 3: Add the public key to your GitHub account
---
  - Log in to your GitHub account.
  - Click on your profile picture or avatar in the top right corner, then select "Settings".
  - Click on "SSH and GPG keys" from the left-hand menu.
  - Click on the "New SSH key" button.
  - Paste the contents of the id_ed25519.pub file into the "Key" field.
  - Give your key a label (e.g., "My MacBook Pro").
  - Click the "Add key" button.

Step 4: Configure your SSH client
---
  - Open the Terminal app on your Mac.
  - Run the command ssh-add -K ~/.ssh/id_ed25519 to add your SSH key to your keychain.

Step 5: Test your SSH connection
---
  - Open a new Terminal window.
  - Run the command ssh -T git@github.com to test your SSH connection.
```
