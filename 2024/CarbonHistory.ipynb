Factors of Global Temperature changes
---

| Factor | Description |
|--------|--------------|
| Greenhouse Gas Emissions | Emissions of carbon dioxide from burning fossil fuels, methane from agriculture and fossil fuel extraction, nitrous oxide from agriculture and industrial processes, and other greenhouse gases like HFCs, PFCs, and SF6 from industrial processes. |
| Deforestation and Land-Use Changes | Deforestation, particularly of tropical rainforests, reduces the planet's ability to absorb and store carbon dioxide. Changes in land use, such as converting forests to agricultural or urban areas, can alter the Earth's surface reflectivity and release stored carbon. |
| Solar Activity and Earth's Orbit | Variations in the amount of solar radiation reaching the Earth due to changes in the Sun's activity and the Earth's orbit around the Sun. |
| Aerosols and Particulate Matter | Aerosols and particulate matter from natural sources (e.g., volcanic eruptions) and human activities (e.g., burning of fossil fuels and biomass) can have a cooling or warming effect on the Earth's surface. |
| Feedback Mechanisms | Positive feedback mechanisms, such as melting of ice and snow, amplify warming by reducing the Earth's reflectivity. Negative feedback mechanisms, like increased cloud cover or vegetation growth, can counteract warming. |
| Natural Climate Variability | Natural fluctuations in the Earth's climate system, such as changes in ocean currents (e.g., El Niño-Southern Oscillation) and atmospheric circulation patterns, can influence global temperatures over shorter timescales. |

1. Greenhouse Effect: CO2 is a potent greenhouse gas that traps heat in the Earth's atmosphere. As CO2 concentrations increase, more heat is retained, leading to an enhanced greenhouse effect and global warming.
2. Anthropogenic Emissions: The rapid increase in atmospheric CO2 levels over the past century is primarily due to human activities, such as burning fossil fuels for energy production, transportation, and industrial processes, as well as deforestation and land-use changes that release stored carbon.
3. Long Atmospheric Lifetime: Once emitted, CO2 molecules can remain in the atmosphere for hundreds to thousands of years, effectively accumulating and prolonging their warming effect over time.
4. Disruption of the Carbon Cycle: Human activities have significantly altered the natural carbon cycle, which regulates the exchange of carbon between the atmosphere, oceans, and terrestrial ecosystems. The excess CO2 emissions have overwhelmed the Earth's natural ability to absorb and sequester carbon, leading to a net increase in atmospheric concentrations.
5. Oceanic Acidification: As the oceans absorb a significant portion of the excess CO2 from the atmosphere, it leads to an increase in acidity (lower pH) of seawater, which can have detrimental effects on marine life and ecosystems.
6.Historical and Projected Impact: Observational data and scientific evidence strongly link the increase in atmospheric CO2 levels since the Industrial Revolution to the observed rise in global temperatures, sea-level rise, and other climate change impacts. Moreover, projections based on future emission scenarios suggest that continued CO2 emissions will further exacerbate these impacts.

| 人為排放 |過去一個世紀大氣中CO2濃度快速上升，主要是由於人類活動，如燃燒化石燃料發電、運輸和工業過程，以及砍伐森林和土地利用改變釋放出儲存的碳。 |
| 長期存留 |一旦排放到大氣中，CO2分子可以在大氣層中存留數百至數千年，有效累積並延長其長期變暖效應。 |
| 擾亂碳循環 | 人類活動顯著改變了自然界中調節大氣、海洋和陸地生態系統之間碳交換的碳循環。過量CO2排放超出了地球自然吸收和固存碳的能力，導致大氣中濃度淨增加。 |
| 海洋酸化 | 由於海洋吸收了大量過剩的大氣CO2，導致海水酸度增加（pH值降低），對海洋生物和生態系統產生不利影響。 |
| 歷史和預測影響 | 觀測數據和科學證據將自工業革命以來大氣CO2濃度上升與觀測到的全球變暖、海平面上升等氣候變化影響緊密聯繫在一起。此外，基於未來排放情景的預測表明，持續的CO2排放將進一步加劇這些影響。 |

雖然其他溫室氣體如甲烷和一氧化二氮也會導致氣候變化，但由於CO2的存在量和人類排放規模，它是最重要且最持久的貢獻因素。減少CO2排放並穩定大氣濃度是緩解和適應全球氣候變化後果的關鍵步驟。

Notes:

1. **Carbon Capture and Storage (CCS)**: This technology captures CO2 emissions from industrial processes or power plants and stores them underground in geological formations. It operates at high temperatures and has a moderate cost compared to other options.

2. **Direct Air Capture (DAC)**: This involves capturing CO2 directly from the ambient air using chemical sorbents. While it operates at ambient temperatures, the cost is currently high, making it less economically viable at present.

3. **Bioenergy with Carbon Capture and Storage (BECCS)**: This combines biomass-based energy production with CCS, capturing and storing the CO2 emissions from the biomass combustion process. It operates at high temperatures and has a relatively high cost.

4. **Carbon Capture and Utilization (CCU)**: This involves capturing CO2 emissions and converting them into valuable products like fuels, chemicals, or building materials. The temperature and cost vary depending on the specific utilization process.

The goals for 2032 are ambitious but necessary to keep the mid-century net-zero target in play, as stated in the reference. These technologies, along with other mitigation strategies, are crucial in reducing global carbon emissions and combating climate change.
