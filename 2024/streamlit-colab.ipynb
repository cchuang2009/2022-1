Any knows AI. Now here, we introduce the famous AI, [Claude AI](claude.ai)  to help coding:
1. make a Hello app
2. create input column, display "Hello $name$ while the visit input ones name.

```python
%%writefile app.py

import streamlit as st

def main():
    # Title
    st.title("Hello App")

    # Name input
    name = st.text_input("Enter your name:", "")

    # Display greeting
    if name:
        greeting = f"Hello {name}"
    else:
        greeting = "Hello everyone"

    st.write(greeting)

if __name__:
   main()
```
