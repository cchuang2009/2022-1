```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
hotdogs = pd.read_csv("VisualizeThis2nd/ch04/data/hot-dog-contest-winners.csv")
```

```python
hotdogs.head(3)
```

```python
# Default barchart
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'])
```

```python
# Year labels
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'], tick_label=hotdogs['Year'])
plt.xticks(rotation=90);
```

```python
# Axis labels
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'])
plt.xlabel("Year")
plt.ylabel("Hot dogs and buns (HDB) eaten")
plt.xticks(rotation=90);
```

```python
# Fill color and border
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'], color="red", edgecolor="none")
plt.xlabel("Year")
plt.ylabel("Hot dogs and buns (HDB) eaten")
```

```python
# Highlight USA colors
fill_colors = ['#821122' if country == 'United States' else '#cccccc' for country in hotdogs['Country']]
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'], color=fill_colors, edgecolor="none")
plt.xlabel("Year")
plt.ylabel("Hot dogs and buns (HDB) eaten")
```

```python
# Highlight record years with color
fill_colors = ['#821122' if record == 1 else '#cccccc' for record in hotdogs['New record']]
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'], color=fill_colors, edgecolor="none")
plt.xlabel("Year")
plt.ylabel("Hot dogs and buns (HDB) eaten")
```

```python
# Spacing
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'], color=fill_colors, edgecolor="none", width=0.3)
plt.xlabel("Year")
plt.ylabel("Hot dogs and buns (HDB) eaten")
plt.title("Nathan's Hot Dog Eating Contest Results, 1980-2010")
```

```python
plt.bar(range(len(hotdogs)), hotdogs['Dogs eaten'], color=fill_colors, edgecolor="none", width=0.5)
plt.xlabel("Year")
plt.ylabel("Hot dogs and buns (HDB) eaten")
```
